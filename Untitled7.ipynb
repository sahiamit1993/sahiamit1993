```python
# Sum of natural numbers up to num

num = 16

if num < 0:
   print("Enter a positive number")
else:
   sum = 0
   # use while loop to iterate until zero
   while(num > 0):
       sum += num
       num -= 1
   print("The sum is", sum)
```

```python

import seaborn as sns
import matplotlib.pyplot as plt

# Load the built-in 'tips' dataset from Seaborn
tips = sns.load_dataset("tips")

# Create a scatter plot with 'total_bill' on the x-axis and 'tip' on the y-axis
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="smoker", size="size")

# Create a line plot with 'total_bill' on the x-axis and 'tip' on the y-axis, grouped by 'sex'
sns.lineplot(data=tips, x="total_bill", y="tip", hue="sex")

# Create a histogram of 'total_bill' with a kernel density estimate
sns.histplot(data=tips, x="total_bill", kde=False)

# Show the plots
plt.show()
```
